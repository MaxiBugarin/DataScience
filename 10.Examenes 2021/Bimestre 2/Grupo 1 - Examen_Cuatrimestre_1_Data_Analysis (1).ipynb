## Transformación de datos 💛

###Nos quedamos con los barrios que tengan más de 5 registros (hospedajes).
Ya que consideramos que tomar como muestra menos de 5 hospedajes por neighbourhood_group, neighbourhood y room_type no representaría el valor general de lo que cuesta hospedarse en esa zona.

```
#Datos 2020 (df2020_super_clean -> m5_grouped_df2020_count)
df2020_clean_g=df2020_super_clean.groupby(['neighbourhood_group',	'neighbourhood',	'room_type']) #agrupamos los registros por esas columnas
grouped_df2020_count=df2020_clean_g.count() #contamos los registros por grupo.

m5_grouped_df2020 = grouped_df2020_count.loc[grouped_df2020_count.price >= 5 ] #nos quedamos con los que ese count >=5

m5_grouped_df2020_count= m5_grouped_df2020.rename(columns={"price": "count"})#renombramos para que aparezca la columna count en lugar de price 
m5_grouped_df2020_count = pd.DataFrame(m5_grouped_df2020_count.loc[:,'count'])#mostramos solo la columna que nos interesa, count
m5_grouped_df2020_count
```

```
#Datos 2021 (df2021_super_clean -> m5_grouped_df2021_count)
df2021_clean_g = df2021_super_clean.groupby(['neighbourhood_group',	'neighbourhood',	'room_type'])
grouped_df2021_count = df2021_clean_g.count()

m5_grouped_df2021 = grouped_df2021_count.loc[grouped_df2021_count.price >= 5 ]

m5_grouped_df2021_count= m5_grouped_df2021.rename(columns={"price": "count"})
m5_grouped_df2021_count = pd.DataFrame(m5_grouped_df2021_count.loc[:,'count'])
m5_grouped_df2021_count

```

###Indexamos DF limpios
Por otra parte, a nuestos DataFrames limpios (**df2020_super_clean** y **df_2021_super_clean**) los indexamos para luego hacer el merge con los DataFrames con las cantidades de registros por grupo (**m5_grouped_df2020_count** y **m5_grouped_df2021_count**)



```
#Datos 2020 (df2020_super_clean -> df2020_clean_keys)
df2020_clean_keys = df2020_super_clean.set_index(['neighbourhood_group','neighbourhood','room_type'])
df2020_clean_keys
```

```
#Datos 2021 (df2021_super_clean -> df2021_clean_keys)
df2021_clean_keys = df2021_super_clean.set_index(['neighbourhood_group','neighbourhood','room_type'])
df2021_clean_keys
```

### Merge para hacer el filtrado
Con un inner "join" combinamos los DataFrames limpios que tienen la data (price, minimum_nights, calculated_host_listings_count) con los DataFrames con la cantidad de registros por grupo (además ya filtrados por count >= 5).

Al hacer inner nos quedamos con los que coincidan según el "on" (en nuestro caso las keys), obteniendo todo el dato pero solo de los neighbourhood_group/neighbourhood/room_type que tienen más de 4 registros)

```
#Datos 2020 (df2020_clean_keys y m5_grouped_df2020_count -> merged_data2020)
merged_data2020 = pd.merge(df2020_clean_keys, m5_grouped_df2020_count, how='inner', on=['neighbourhood_group','neighbourhood','room_type'])
merged_data2020
```

```
#Datos 2021 (df2021_clean_keys y m5_grouped_df2021_count -> merged_data2021)
merged_data2021 = pd.merge(df2021_clean_keys, m5_grouped_df2021_count, how='inner', on=['neighbourhood_group','neighbourhood','room_type'])
merged_data2021
```

El sexto decil (100 a 120 dólares) disminuyó su cantidad de listings en casi un 20% 
El decil 7 (120 a 150 dólares) sufrió un incremento en la cantidad de sus listings de un poco más del 10% (pasó de 2.294 propiedades a 2.527). 
Por último, el octavo decil (150 a 190 dólares) sufrió una caída de 2068 listings en 2020 a 1685 en 2021 (casi 19%).
Concluimos de esta parte del análisis que, consistentente con lo visto anteriormente, la redistribución de los precios no es lo suficientemente uniforme como para poder determinar un alza o caída en los precios ni tampoco en la oferta de los distintos tipos de alojamientos.

## Conclusión
Como se pudo ver anteriormente, los precios hasta abril de 2021 (último mes incluido en el dataset) bajaron muy levemente con respecto al mismo periodo en 2020, un descenso que podría considerarse no significativo.

Nuestra hipótesis es que, de la mano del arribo de la temporada alta, de la baja de casos de coronavirus (apoyada por la vacunación masiva) y de las medidas para la reactivación turística, los precios de Airbnb subirán; pero, para hacer un análisis más exhaustivo y en consecuencia más fiel, sería conveniente contar con los datos de los listings para los meses de mayo, junio, julio y agosto (meses en los que la temperatura empieza a elevarse, los casos bajan y en los que probablemente haya empezado una reactivación económica).
