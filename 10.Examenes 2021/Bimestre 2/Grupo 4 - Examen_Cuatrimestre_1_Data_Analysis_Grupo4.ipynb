###Basado en el uso de percentiles generamos 5 segmentos de alojamientos en función de su precio.

Los criterios utilizados fueron cuartiles, pero sumando una categoría VIP la cual representa el 1% superior de la población.


```
q1 = np.quantile(df_2020['price'], 0.25)
q2 = np.quantile(df_2020['price'], 0.50)
q3 = np.quantile(df_2020['price'], 0.75)
vip = np.quantile(df_2020['price'], 0.99)
max = np.max(df_2020['price'])

bins = [0, q1, q2, q3, vip, max]

labels = ['Classic','Gold','Platinum','Black','Black VIP']

df_2020['Grupo_precio'] = pd.cut(df_2020['price'], bins, labels=labels)

df_2021['Grupo_precio'] = pd.cut(df_2021['price'], bins, labels=labels)
```

```
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```
ny_mask = np.array(Image.open('/content/mascara.png'))
```

```
# Transform your mask into a new one that will work with the function:
transformed_ny_mask = np.ndarray((ny_mask.shape[0],ny_mask.shape[1]), np.int32)

for i in range(len(ny_mask)):
    transformed_ny_mask[i] = list(map(transform_format, ny_mask[i]))

wc = WordCloud(background_color="white", max_words=1000, mask=transformed_ny_mask,
               stopwords=stopwords, contour_width=1, contour_color='black')

# Generate a wordcloud
wc.generate(text)

# store to file
#wc.to_file("img/wine.png")

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## Conclusión

* Como conclusión preliminar desde el análisis para limpieza de datos, podemos ver que en **2020 había 44.666 propiedades** ofrecidas en alquiler temporario mientras que en **2021 había 36.905**. Eso representa una baja de 17,37% y puede ser atribuido fácilmente a la pandemia y las restricciones impuestas al turismo, sumado a las precauciones de los propietarios.

* Otra conclusión basada en la visualización "Cantidad y tipo de inmueble por distrito" es que **la cantidad y opción predominante en ambos años es el departamento entero en Manhattan, seguido por Brooklyn**, aunque en este último hay mayor proporción en la elección de cuartos privados y esto posiblemente se debe a que al ser una opción mas económica es elegida por gente trabajando en Nueva York mas que por turismo.

* Por otra parte, la visualización denominada **"Densidad y precios por grupo de barrios" nos muestra que en ambos años la densidad y distribución es similar**, distinguiendose que tanto en Queens y Bronx la mayor cantidad de propiedades son las que rondan el precio de los 50 dólares (la parte mas ancha del violín) siendo que son los barrios mas pobres y menos demandados es lógica la conclusión, en tanto que en Manhattan está mas cerca de los 100 dólares y la mediana (el punto blanco en el centro del violín) mas arriba incluso. 

* Las visualizaciones de catplot ilustran esta situación mostrando la distribución por tipo de propiedad **notándose nuevamente una predominación de los departamentos enteros**  salvo en los distritos mas económicos donde también se nota que casi no hay oferta de cuartos de hoteles.
