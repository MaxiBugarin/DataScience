### Funciones para lectura de los archivos

Se crearon una serie de funciones que lee tanto el archivo de nacimientos como el archivo de expectativa de vida de los distintos distritos de Barcelona.

```python
def leer_archivo_csv_nacimientos(archivo_csv, diccionario):
  """
  Función para leer el archivo csv de nacimientos.
  
  Normalizamos los valores de "Girls" y "Boys" a "Female" y "Male" respectivamente.
  Generamos un diccionario con todos los datos de los distritos, por año y por sexo.   {'id_distrito': nacimientos:{'año':{'female': x, 'male': x}}}

  Parametros:
    archivo_csv >>> es la ruta del archivo csv
    diccionario >>> es el diccionario donde voy a guardar los datos
  Return: no devuelvo nada
  """
  try:
    with open(archivo_csv, mode='r') as csv_file:
      csv_reader = csv.DictReader(csv_file, delimiter=';') # Delimitador ;
      for row in csv_reader:      # itero por linea del csv
          Male    = int(diccionario.get(row["District Code"],{}).get('nacimientos',{}).get(row["Year"],{}).get("Male",0))
          Female  = int(diccionario.get(row["District Code"],{}).get('nacimientos',{}).get(row["Year"],{}).get("Female",0))

          if row['Gender'] == 'Boys':     # Normalización de valores "Boys" a "Male"
            Male += int(row["Number"])
          elif row['Gender'] == 'Girls':  # Normalizacion de valores "Girls" a "Female"
            Female += int(row["Number"])
          
          nacimientos = diccionario.get(row["District Code"],{}).get('nacimientos',{})
          nacimientos[row['Year']] = {'Male':Male, 'Female':Female}
          diccionario[row["District Code"]] = {'nacimientos': nacimientos}    # Cargo el diccionario de nacimientos
  except FileNotFoundError:
    print("Inicializar el archivo")

leer_archivo_csv_nacimientos(archivo_nacimientos, barcelona)  # Me cargo los datos de los nacimientos.
#print(json.dumps(barcelona,indent=4, ensure_ascii=False))
```

```python
def leer_archivo_csv_distritos(archivo_csv, diccionario):
  """
  Función para leer el archivo csv de nacimientos para obtener los nombres de los distritos.

  Generamos un diccionario con todas las descripciones de los distritos. {'id_distrito': 'descripcion'}

  Parametros:
    archivo_csv >>> es la ruta del archivo csv
    diccionario >>> es el diccionario donde voy a guardar los datos
  Return: no devuelvo nada
  """
  try:
    with open(archivo_csv, mode='r') as csv_file:
      csv_reader = csv.DictReader(csv_file, delimiter=';') # Delimitador ;
      for row in csv_reader:      # itero por linea del csv
          diccionario[row["District Code"]] = row["District Name"]  # Cargo el diccionario de descripciones de nombres de distritos
  except FileNotFoundError:
    print("Inicializar el archivo")

leer_archivo_csv_distritos(archivo_nacimientos, distritos)    # Me cargo en el diccionario los nombres de los distritos.
#print(json.dumps(distritos,indent=4, ensure_ascii=False))
```

```python
def leer_archivo_csv_poblacion(archivo_csv, diccionario):
  """
  Función leer el archivo csv de poblacion.
  
  Generamos un diccionario con todos los datos de población de los distritos, por año y por sexo.   {'id_distrito': poblacion:{'año':{'Female': x, 'Male': x, 'Total': x}}}

  Parametros:
    diccionario >> es el diccionario general con datos de barcelona
    archivo_csv >>> es la ruta del archivo csv
  Return: diccionario con datos de poblacion por año y distritro
  """
  dic_poblacion = {}
  try:
    with open(archivo_csv, mode='r', encoding = "ISO-8859-1") as csv_file:
      csv_reader = csv.DictReader(csv_file, delimiter=';') # Delimitador ;
      for row in csv_reader:      # itero por linea del csv
        Male = row['male']
        Female = row['female']
        Total = row['total']
        Año = row['anio']
        id_distrito = str(row['id_distrito'])
        poblacion = dic_poblacion.get(id_distrito,{}).get('poblacion',{})
        poblacion[Año] = {'Female': Female, 'Male': Male, 'Total': Total}
        dic_poblacion[id_distrito] = {'poblacion': poblacion}

    for id_distrito in dic_poblacion.keys():
      diccionario[id_distrito]['poblacion'] = dic_poblacion[id_distrito].get('poblacion',{})
  except FileNotFoundError:
    print("Inicializar el archivo")

leer_archivo_csv_poblacion(archivo_poblacion, barcelona)  # Me cargo los datos de la poblacion.
#print(json.dumps(barcelona,indent=4, ensure_ascii=False))
```

```python
def leer_archivo_csv_defunciones(archivo_csv, diccionario):
  """
  Función para leer el archivo csv de defunciones.
  
  Generamos un diccionario con todos los datos de defunciones de los distritos, por año y por sexo.   {'id_distrito': defunciones:{'año':{'Female': x, 'Male': x, 'Total': x}}}

  Parametros:
    diccionario >> es el diccionario general con datos de barcelona
    archivo_csv >>> es la ruta del archivo csv
  Return: actualizo el diccionario de con datos de barcelona con datos de defunciones por año y distritro
  """
  dic_defunciones = {}
  try:
    with open(archivo_csv, mode='r', encoding = "ISO-8859-1") as csv_file:
      csv_reader = csv.DictReader(csv_file, delimiter=';') # Delimitador ;
      for row in csv_reader:      # itero por linea del csv
        Male = row['male']
        Female = row['female']
        Total = row['total']
        Año = row['anio']
        id_distrito = str(row['id_distrito'])
        defunciones = dic_defunciones.get(id_distrito,{}).get('defunciones',{})
        defunciones[Año] = {'Female': Female, 'Male': Male, 'Total': Total}
        dic_defunciones[id_distrito] = {'defunciones': defunciones}

    for id_distrito in dic_defunciones.keys():
      diccionario[str(id_distrito)]['defunciones'] = dic_defunciones[id_distrito].get('defunciones',{})
  except FileNotFoundError:
    print("Inicializar el archivo")

leer_archivo_csv_defunciones(archivo_defunciones, barcelona)  # Me cargo los datos de las defunciones.
#print(json.dumps(barcelona,indent=4, ensure_ascii=False))
```

```python
def promedioPorRango(Sumas):
  """
  Función para calcular los promedios por sexo de los periodos que existen de cada distrito.

  Parametros:
    Sumas >>> diccionario temporal de sumas de expectativa de vida por periodo y sexo de cada distrito
  Return:
    PromedioList >>> devuelve el diccionario de promedio actualizado
  """
  for key, value in Sumas.items():
    for i in range(0,5):
        if Sumas[key]['FemaleCuenta'][i] != 0:
            PromediosList[key]['Female'][i] = Sumas[key]['FemaleSuma'][i] / Sumas[key]['FemaleCuenta'][i]

        if Sumas[key]['MaleCuenta'][i] != 0:
            PromediosList[key]['Male'][i] = Sumas[key]['MaleSuma'][i] / Sumas[key]['MaleCuenta'][i]

  return(PromediosList)

PromediosList = promedioPorRango(Sumas) # Calculo los promedio de cada uno de los periodos por sexo de cada distrito
#print(json.dumps(PromediosList,indent=4, ensure_ascii=False))
```

```python
def ordenar_diccionario(diccionario, orden=False):
  """
  Función para ordenar un diccionario clave, valor de forma ascendente o descendente

  Parametros:
    diccionario >>> es el diccionario a ordenar con formato clave: valor
    orden >>> es la forma de ordenamiento False (A->Z) o True (Z->A) 
  Return:
    diccionario_ordenado >>> es el mismo diccionario pero ordendo
  """
  diccionario_ordenado = {k: v for k, v in sorted(diccionario.items(), key=lambda item: item[1], reverse= orden)}

  return diccionario_ordenado
```

```python
def expectativa_vida_distrito(sexo, diccionario, orden=True):
  """
  Función generica para obtener la expectativa de vida por sexo.

  Parametros:
    sexo >>> descripcion del sexo (Female|Male|Total)
    diccionario >>> es el diccionario con la expectativa de vida
    orden >>> para ordenar la salida
  Return:
    vida >>> diccionario con la expectativa de vida ordenada por distrito segun el parametro de sexo
  """
  vida = {}

  for key, value in diccionario.items():
      vida[key] = diccionario[key][sexo]

  vida = ordenar_diccionario(vida, orden)

  return vida
```

```python
def calcular_natalidad(diccionario_nacimientos):
  """
  Calculo de totales generico para calculo de natalidad por distrito

  Parametros: 
    nombre_campo >>> es el campo donde realizo el agrupamiento
    diccionario_ventas >>> es el diccionario que contiene toda la data
  Return: devuelvo un diccionario con el siguiente formato:  { "id_distrito": nacimientos_totales }
  """
  dic = {}
  for distrito, años in diccionario_nacimientos.items():
    for k,año in diccionario_nacimientos.get(distrito)['nacimientos'].items():
      dic[distrito] =  año.get('Male',0) + año.get('Female',0) + dic.get(distrito,0)
  return dic
```

```python
def diccionario_cantidad(diccionario, cantidad):
  """
  Funcion para dejar X cantidad de claves en el diccionario.

  Parametros:
    diccionario >>> es el diccionario a cortar la cantidad de items
    cantidad >>> es el valor de items a dejar
  Return:
    diccionario_cantidad >>> es el diccionario cortado la cantidad de items
  """
  diccionario_cantidad = {k: diccionario[k] for k in list(diccionario)[-cantidad:]}  # Filtro diccionario primero 10 key
  return diccionario_cantidad
```

```python
def diccionario_campo(nombre_campo, diccionario):
  """
  Función para obtener datos extras

  Genera un diccionario con el dato extra que le pidamos por id_distrito  {'id_distrito': x}
  
  Parametros:
    nombre_campo >>> es el nombre del campo del dato extra (m2|poblacion|renta_media|enfermedades|densidad_poblacion_km2|superficie_km2|tipo_distrito|nombre_español)
    diccionario >>> es el diccionario de datos extras
  Return:
    diccionario con el valor del atributo pedido por distrito
  """
  dic = {}
  for x in range(1,11):
    dic[str(x)] = diccionario[str(x)].get(nombre_campo)

  return dic
```

```python
def calcular_ratios(dic_numerador, dic_denominador):
  """
  Calculo de ratios

  Parametros: 
    dic_numerador >>> es el diccionario que se va a utilizar para el numerador
    dic_denominador >>> es el diccionario que se va a utilizar para el denominador
  Return: devuelvo un diccionario con el siguiente formato:  { "id_distrito": ratio }
  """
  dic = {}
  for id_distrito, valor_denominador in dic_denominador.items():
      dic[id_distrito] =  dic_numerador[id_distrito] / valor_denominador * 100
  return dic
```

```python
def calcular_tasas(diccionario):
  """
  Calculo de ratios tasa natalidad y tasa de mortalidad

  Parametros: 
    diccionario >>> es el diccionario que se va a utilizar para calcular las tasas
  Return: no de vuelve nada
  """
  ratio_nac = {}
  ratio_def = {}
  tasas = {}

  for id_distrito in barcelona.keys():
    anio1 = {}
    anio2 = {}
    for año, item in barcelona[id_distrito].get('nacimientos').items():
      n = item['Male']+item['Female']
      d = int(barcelona[id_distrito]['defunciones'][año].get('Total'))
      p = int(barcelona[id_distrito]['poblacion'][año].get('Total'))

      anio = ratio_nac.get(id_distrito,{}).get('años',{})
      anio1[año] = n/p*1000                                   # Tasa de natalidad
      ratio_nac[id_distrito] = {'años': anio1}

      anio2[año] = d/p*1000                                   # Tasa de mortalidad
      ratio_def[id_distrito] = {'años': anio2}
    
    tasa1 = []
    for k, v in ratio_nac[id_distrito].items():
      for k1, a in v.items():
        tasa1.append(a)
    ratio_nac[id_distrito]['tasa_natalidad'] = np.average(tasa1)

    tasa2 = []
    for k, v in ratio_def[id_distrito].items():
      for k1, a in v.items():
        tasa1.append(a)
    ratio_nac[id_distrito]['tasa_mortalidad'] = np.average(tasa1)
    ratio_nac[id_distrito].pop('años')

  for id_distrito in diccionario.keys():  # Incorporo al diccionario los valores de tasa natalidad y mortalidad
    diccionario[id_distrito]['tasa_natalidad'] = ratio_nac[id_distrito]['tasa_natalidad']
    diccionario[id_distrito]['tasa_mortalidad'] = ratio_nac[id_distrito]['tasa_mortalidad']

calcular_tasas(barcelona) # Calculo las tasas de natalidad y mortalidad
#print('tasa_natalidad ', barcelona['1']['tasa_natalidad'])
#print('tasa_mortalidad ', barcelona['1']['tasa_mortalidad'])
```

```python
def graficar_(dic_grafico, titulo_grafico, x_label, y_label, colores, mostrar_texto=True, linea=False, barra=False, horizontal=False, linea_size=0, alpha_value=None, size_x=12, size_y=5, redondear=False, decimales=None):
  """
  Graficamos el diccionario que se recibe por parametro en modo por default Fill_between

  Podemos graficar graficos de lineas, barras, barras horizontal y una serie de parametros para personalizarlos

  Parametros:
    dic_grafico >>> diccionario a grafica con formato clave, valor
    titulo_grafico >>> descripcion del titulo del grafico
    x_label >>> descripcion del label del eje X
    y_label >>> descripcion del label del eje Y 
    colores >>> colores a utilizar puede ser un solo color o una lista de colores
    mostrar_texto >>> es para mostrar el texto dentro del grafico (True|False). Default=True
    linea >>> para activar el modo de grafico linea (True|False) Default=False
    barra >>> para activar el modo de grafico barra (True|False) Default=False
    horizontal >>> para activar el modo de grafico horizontal en conjunto con el parametro "barra" (True|False) Default=False
    linea_size >>> para setear el valor de grosos de la linea. Default=0
    alpha_value >>> valor de transparencia para grafico fill_between
    size_x >>> valor del tamaño del grafico en eje X
    size_y >>> valor del tamaño del grafico en eje Y
  Return:
    No devuelve nada solo imprime
  """
  fig = plt.figure(figsize=(size_x, size_y)) 
  ax = fig.subplots()
  x_data = []
  y_data = []

  for k, values in dic_grafico.items(): 
    x_data.append(str(distritos[k]))              # Agregamos la data del eje X
    if redondear == True:
      y_data.append(round(values, decimales))     # Agregamos la data del eje Y
    else:
      y_data.append(values)                       # Agregamos la data del eje Y

  if mostrar_texto ==  True:
    for index, value in enumerate(y_data):  # Texto en barras
        if barra == True and horizontal==True:
          plt.text(value, index, str(value))
        else:
          plt.text(index, value, str(value), ha='center', va='bottom')

  ax.set_title(titulo_grafico, loc = "center", fontdict = {'fontsize':15, 'color':'tab:grey'})  # Titulo del grafico
  ax.set_ylabel(y_label, fontdict = {'fontsize':12})                    # Label del eje Y
  ax.set_xlabel(x_label, fontdict = {'fontsize':12})                    # Label del eje X
  
  plt.xticks(rotation=45) # Rotacion de los labels del eje X

  if horizontal == True:
    ax.xaxis.grid(color='#e9ecef', linestyle='dashdot', linewidth=1) # Seteo de lineas en eje X
  else:
    ax.yaxis.grid(color='#e9ecef', linestyle='dashdot', linewidth=1) # Seteo de lineas en eje Y

  ax.set_axisbelow(True)  # Lineas eje atras el grafico

  if barra == False and linea == False:
    ax.plot(x_data, y_data, color=colores, linewidth=linea_size)        # Establecemos los parametros del grafico si es Linea
    ax.fill_between(x_data, y_data, color=colores, alpha=alpha_value)   # Establecemos los parametros del grafico si es linea fill
  if linea == True:
    if linea_size == 0:
      linea_size = 1
    ax.plot(x_data, y_data, color=colores, linewidth=linea_size)  # Establecemos los parametros del grafico si es Linea
  if barra == True and horizontal==False:
    ax.bar(x_data, y_data, color=colores)           # Establecemos los parametros del grafico si es Barra y Vertical
  if barra == True and horizontal==True:
    ax.barh(x_data, y_data, color=colores)          # Establecemos los parametros del grafico si es Barra y Horizontal
  plt.savefig(titulo_grafico)                       # Guardo la imagen del grafico
  plt.show()                                        # Ploteamos
```

```python
def inicializar_mapa(archivo_mapa):
  """
  Función para inicializar el diccionario de mapa

  Parametros:
    archivo_mapa >>> es el nombre del archivo que contiene la información de los puntos del mapa
  Return:
    diccionario con los puntos del mapa que corresponden a cada distrito  {'id_distrito': [[x,y],[x,y]]}
  """
  polygon = {}
  with open(archivo_mapa) as f:
    polygon = json.load(f)

  return polygon
  
def graficar_mapa(map, titulo_grafico, colores, colores_texto, dic, ver_texto=False):
  """
  Función para graficar el mapa

  Permite recibir un diccionario generico y visualizarlos en el mapa

  Parametros:
    map >>> es el diccionario que contiene los datos de puntos para cada distrito
    titulo_grafico >>> es la descripcion del titulo del grafico
    colores >>> es una lista con la serie de colores a utilizar en las areas del mapa
    colores_texto >>> es una lista con la serie de colores a utilizar en los texto
    dic >>> es el diccionario generico {'id_distrito': x}
    ver_texto >>> para ver o no el texto arriba del area (False|True)
  Return:
    Imprime por pantalla el mapa
  """
  fig = plt.figure(figsize=(8, 8))
  ax = fig.subplots()

  for id_distrito in dic.keys():
    x = map[str(id_distrito)].get('x')
    y = map[str(id_distrito)].get('y')
    label_texo = distritos[id_distrito] + ' ' + str(int(dic[id_distrito]))
    ax.fill(x, y , color = colores[list(dic.keys()).index(id_distrito)], label = label_texo)
    
    if ver_texto == True:       # Agrego los labels en el mapa
      min_x = np.array(x).min()
      max_x = np.array(x).max()
      min_y = np.array(y).min()
      max_y = np.array(y).max()

      offset_x, offset_y = parametros_offset_mapa(id_distrito) # Para calculo de desplazamiento de texto arriba del grafico

      xx = min_x + (max_x-min_x)/2 + offset_x
      yy = min_y + (max_y-min_y)/2 + offset_y
      label = distritos[str(id_distrito)].replace(' ','\n')
      label_color = colores_texto[list(dic.keys()).index(id_distrito)]
      plt.text(xx, yy, label, color = label_color)
  
  plt.legend(loc='lower center', ncol=1, bbox_to_anchor=(0.8, 0., 0.5, 0.5))
  plt.title(titulo_grafico)                         # Titulo del grafico
  plt.axis('off')                                   # Oculto los ejes
  plt.savefig(titulo_grafico)                       # Guardo la imagen del grafico
  plt.show()                                        # Ploteamos le grafico

def parametros_offset_mapa(id_distrito):
  """
  Calcula el offset para label arriba del grafico.

  Las areas de los distritos no son uniformes y el centro no siempre entra el label
  Parametro:
    id_distrito >>> codigo de distrito
  Return:
    offset_x >>> valor a desplazar en eje X para la correccion
    offset_y >>> valor a desplazar en eje Y para la correccion
  """
  offset_x = 0
  offset_y = 0
  if int(id_distrito) == 1:
    offset_x = -0.006
    offset_y = 0.005
  elif int(id_distrito) == 2:
    offset_x = -0.006
    offset_y = -0.002   
  elif int(id_distrito) == 3:
    offset_x = -0.002
    offset_y = 0.002
  elif int(id_distrito) == 4:
    offset_x = -0.005
    offset_y = -0.006
  elif int(id_distrito) == 5:
    offset_x = -0.008
    offset_y = -0.004
  elif int(id_distrito) == 7:
    offset_x = -0.02
    offset_y = 0.004
  else:
    offset_y = -0.0015

  return offset_x, offset_y
```

```python
def grafico_versus(map, titulo_grafico, titulo_grafico1, titulo_grafico2, color1, color2, colores_texto, dic1, dic2, ver_texto=True, redondear1=None, redondear2=None):
  """
  Función para graficar dos mapa para compararlos segun dos metricas

  Parametros:
    map >>> es el diccionario que contiene los datos de puntos para cada distrito
    titulo_grafico >>> es el titulo general del grafico
    titulo_grafico1 >>> es el titulo del mapa 1
    titulo_grafico2 >>> es el titulo del mapa 2
    color1 >>> es la lista de colores para el mapa1
    color2 >>> es la lista de colores para el mapa2
    colores_texto >>> es la lista de colores para los texto arriba del area del mapa
    di1c >>> es el diccionario generico para el mapa1 {'id_distrito': x}
    dic2 >>> es el diccionario generico para el mapa2 {'id_distrito': x}
    ver_texto >>> para ver o no el texto arriba del area (False|True) 
  """
  fig, axes = plt.subplots(1, 2, figsize=(12,6.5))  
  fig.tight_layout(pad=4.0)

  # Completo el mapa 1 con sus datos
  for id_distrito in dic1.keys():
    x = map[str(id_distrito)].get('x')
    y = map[str(id_distrito)].get('y')
    label_texo = distritos[id_distrito] + ' ' + str(round(dic1[id_distrito], redondear1))
    axes[0].fill(x, y , color = color1[list(dic1.keys()).index(id_distrito)], label = label_texo)
    ######
    if ver_texto == True:   # Seccion para agregar el texto si es necesario
      # Calculo de maximos y minimos
      min_x = np.array(x).min()
      max_x = np.array(x).max()
      min_y = np.array(y).min()
      max_y = np.array(y).max()

      # Calculo de offset
      offset_x, offset_y = parametros_offset_mapa(id_distrito) # Para calculo de desplazamiento de texto arriba del grafico

      # Calculo de centro para labels
      xx = min_x + (max_x-min_x)/2 + offset_x
      yy = min_y + (max_y-min_y)/2 + offset_y
      label = distritos[(id_distrito)].replace(' ','\n')
      label_color = colores_texto[list(dic1.keys()).index(id_distrito)]
      axes[0].text(xx, yy, label, color = label_color)  # Se agregan los labels en el centro del area del grafico por distrito
    ######

  # Completo el mapa 2 con sus datos
  for id_distrito in dic2.keys():
    x = map[str(id_distrito)].get('x')
    y = map[str(id_distrito)].get('y')
    label_texo = distritos[id_distrito] + ' ' + str(round(dic2[id_distrito], redondear2))
    axes[1].fill(x, y , color = color2[list(dic2.keys()).index(id_distrito)], label = label_texo)
    ######
    if ver_texto == True:  # Seccion para agregar el texto si es necesario
      # Calculo de maximos y minimos
      min_x = np.array(x).min()
      max_x = np.array(x).max()
      min_y = np.array(y).min()
      max_y = np.array(y).max()

      # Calculo de offset
      offset_x, offset_y = parametros_offset_mapa(id_distrito) # Para calculo de desplazamiento de texto arriba del grafico

      # Calculo de centro para labels
      xx = min_x + (max_x-min_x)/2 + offset_x
      yy = min_y + (max_y-min_y)/2 + offset_y
      label = distritos[(id_distrito)].replace(' ','\n')
      label_color = colores_texto[list(dic2.keys()).index(id_distrito)]
      axes[1].text(xx, yy, label, color = label_color) # Se agregan los labels en el centro del area del grafico por distrito
    ######

  axes[0].set_title(titulo_grafico1)  # Titulo para grafico 1
  axes[1].set_title(titulo_grafico2)  # Titulo para grafico 2

  axes[0].axis('off') # Se ocultan los ejes del mapa 1
  axes[1].axis('off') # Se ocultan los ejes del mapa 2
  axes[0].legend(loc='lower center', ncol=1, bbox_to_anchor=(0.6, 0., 0.9, 1))  # Se setea la posicion de las leyendas grafico 2
  axes[1].legend(loc='lower center', ncol=1, bbox_to_anchor=(0.6, 0., 0.9, 1))  # Se setea la posicion de las leyendas grafico 2

  fig.suptitle(titulo_grafico, size=20, y=1)   # Titulo del grafico
  plt.savefig(titulo_grafico)                  # Guardo la imagen del grafico
  plt.show()                                   # Ploteamos el grafico versus
```

## Persistencia (Almacenamiento)

Se realizo la generación de una función generica para guardar los distintos diccionarios utilizados en el anális.

```python
def grabar_archivo_json(nombre_archivo, diccionario, headers):
  """
  Graba el archivo pasado por parametro en formato json

  Guarda el diccionario pasado por parametro en el archivo

  Parametros:
    nombre_archivo >>> es la ruta del archivo jscon a guardar
    diccionario >>> es el diccionario que tiene los datos a guardar
  Return: no devuelvo nada
  """
  with open(nombre_archivo, mode='w') as csv_file:    # utilizo la variable global del nombre del archivo
    fieldnames = headers                              # Definio la cabecera del csv
    writer = csv.DictWriter(csv_file, fieldnames = fieldnames)

    writer.writeheader()
    for key, values in diccionario.items():                               # Recorro el diccionario
      writer.writerow({headers[0]: str(key), headers[1]: str(values)})    # Grabo la fila con los datos
```
